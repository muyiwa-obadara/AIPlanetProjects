### Proportion of passenger who survived.

```python
surv_prop = titanic_data.Survived.sum()/titanic_data.Survived.count()
surv_prop.round(2)
```

#### Median Fare of Passengers

```python
median_fare = titanic_data.Fare.median()
median_fare.round(4)
```

#### Survival by Sex

```python
female_data = titanic_data[titanic_data.Sex == "female"]
male_data = titanic_data[titanic_data.Sex == "male"]

survived_female = female_data.Survived == 1
print("Percentage of female that survived: {}%".format(round(survived_female.sum()*100/len(female_data),2)))

survived_male = male_data.Survived == 1
print("Percentage of male that survived: {}%".format(round(survived_male.sum()*100/len(male_data),2)))
```

Considering the sex, it is seen that out of the female gender, more female survived than the male gender.

#### Survival by Pclass

```python
class_1 = titanic_data[titanic_data.Pclass == 1]
class_2 = titanic_data[titanic_data.Pclass == 2]
class_3 = titanic_data[titanic_data.Pclass == 3]

class_1_survived = (class_1.Survived == 1).sum()
class_2_survived = (class_1.Survived == 1).sum()
class_3_survived = (class_1.Survived == 1).sum()

print("Percentage of passengers who survived from First Class Cabin: {}%".format(round((class_1_survived*100/len(class_1)), 2)))
print("Percentage of passengers who survived from First Class Cabin: {}%".format(round((class_2_survived*100/len(class_2)), 2)))
print("Percentage of passengers who survived from First Class Cabin: {}%".format(round((class_3_survived*100/len(class_3)), 2)))
```
